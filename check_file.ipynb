```python
import glob

# Lấy tất cả file parquet
file_list = glob.glob("mobisureadmin_GLM_HEALTH_AC*.parquet")

schemas = {}
for f in file_list:
    df = pd.read_parquet(f)
    schemas[f] = list(df.columns)

# Lấy schema chuẩn (của file đầu tiên)
base_schema = schemas[file_list[0]]

print("Schema chuẩn (theo file đầu tiên):")
print(base_schema)

# So sánh các file khác
print("\nSo sánh schema:")
for f, cols in schemas.items():
    if cols != base_schema:
        print(f"⚠️ File {f} KHÁC schema")
        # In ra cột bị thiếu / thừa
        missing = set(base_schema) - set(cols)
        extra   = set(cols) - set(base_schema)
        if missing:
            print("  Thiếu cột:", missing)
        if extra:
            print("  Thừa cột:", extra)
    else:
        print(f"✅ File {f} OK (schema giống)")
```
